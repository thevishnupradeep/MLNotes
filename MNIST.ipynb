# Basic MNIST Neural Network

```python
import torch
import torchvision.datasets as datasets 
import torchvision.transforms as transforms
import torch.nn as nn
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

## Load Data

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,)),])
```

```python
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
train_loader = DataLoader(mnist_trainset, batch_size=10, shuffle=True)
```

```python
mnist_testset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)
test_loader = DataLoader(mnist_testset, batch_size=10, shuffle=True)
```

## Data Exploration

```python
for data in train_loader:
    print(data[0].shape, data[1].shape)
    break
```

## Neural Network

```python
class Network(nn.Module):
    def __init__(self):
        super(Network, self).__init__()

        self.linearA = nn.Linear(28*28, 128)
        self.linearB = nn.Linear(128, 64)
        self.linearC = nn.Linear(64, 32)
        self.linearD = nn.Linear(32, 16)
        self.linearE = nn.Linear(16, 10)

        self.relu = nn.ReLU()

    def forward(self, img):
        x = img.view(-1, 28*28)

        x = self.relu(self.linearA(x))
        x = self.relu(self.linearB(x))
        x = self.relu(self.linearC(x))
        x = self.relu(self.linearD(x))
        x = self.linearE(x)
        
        return x
```

```python
net = Network()
```

## Text

```python
cross_el = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.001) #e-1
epoch = 10
```

```python
for epoch in range(epoch):
    net.train()

    for data in train_loader:
        x, y = data
        optimizer.zero_grad()
        output = net(x.view(-1, 28*28))
        loss = cross_el(output, y)
        loss.backward()
        optimizer.step()
```
